# NBA Data Exploration

This notebook explores the NBA data from Sportradar MCP and analyzes team/player statistics.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.insert(0, '..')

from src.utils.database import get_session, Team, Game, Player, PlayerStats, TeamStats
from src.data.sportradar_client import sportradar_client
```

## Load Teams

```python
session = get_session()
teams = session.query(Team).all()
print(f"Total teams: {len(teams)}")
teams_df = pd.read_sql(session.query(Team).statement, session.bind)
teams_df.head()
```

## Load Games

```python
games = session.query(Game).limit(100).all()
games_df = pd.read_sql(session.query(Game).statement, session.bind)
print(f"Total games in database: {len(games_df)}")
games_df.head()
```

## Team Statistics Distribution

```python
team_stats = pd.read_sql(session.query(TeamStats).statement, session.bind)
print(f"Total team stat records: {len(team_stats)}")
team_stats.describe()
```

## Player Statistics Analysis

```python
player_stats = pd.read_sql(session.query(PlayerStats).statement, session.bind)
print(f"Total player stat records: {len(player_stats)}")

key_stats = ['points', 'rebounds', 'assists', 'minutes', 'steals', 'blocks']
player_stats[key_stats].describe()
```

## Top Scorers

```python
avg_points = player_stats.groupby('player_id')['points'].mean().sort_values(ascending=False)
top_scorers = avg_points.head(20)
plt.figure(figsize=(12, 6))
top_scorers.plot(kind='bar')
plt.title('Top 20 Average Scorers')
plt.xlabel('Player ID')
plt.ylabel('Average Points')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Net Rating Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(team_stats['net_rating'].dropna(), bins=30)
plt.title('Distribution of Net Ratings')
plt.xlabel('Net Rating')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## Pace Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(team_stats['pace'].dropna(), bins=30)
plt.title('Distribution of Pace')
plt.xlabel('Pace')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## Key Findings

- Team and player statistics loaded successfully
- Net rating follows a roughly normal distribution centered around 0
- Pace varies significantly across games
